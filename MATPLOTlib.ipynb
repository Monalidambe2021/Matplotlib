# Matplotlib Tutorial
Matplitlib is a plotting library for the Python programming language and its numerical mathamatics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter, wxPython, Qt, or GTK+.
Some of the important properties of matpltlib are:
1. Generally easy to get started for simple plots.
2. Support for custom labels and texts.
3. Great control of every element in a figure.
4. High-quality output in many formats.
5. Very customizable in general.

# Sine Wave

```python
# compute the x and y coordinates for points on a sine curve
x= np.arange(0,4* np.pi, 0.1)
y= np.sin(x)
plt.title('sine wave form')
#plot the points using matplotlib
plt.plot(x,y)
plt.show()
```

# Sine and Cosine curve

```python
#subplot()
# compute the xand y coordinates for points on sine and cosine curve
x= np.arange(0,5*np.pi,0.1)
y_sin= np.sin(x)
y_cos= np.cos(x)

## set up a subplot grid that has height 2 and width 1,
##and set the first such subplot as active

plt.subplot(2,1,1)

## make the first plot
plt.plot(x,y_sin, 'g--')
plt.title('Sine')

# set the second subplot as active, and make the second plot
plt.subplot(2,1,2)
plt.plot(x,y_cos,'r--')
plt.title('Cosine')

## show the figure
plt.show()
```

# Bar plot

```python
## Bar plot
x= [2,8,10]
y=[11,16,9]

x2=[3,9,11]
y2=[6,15,7]
plt.bar(x,y)
plt.bar(x2,y2, color= 'r')
plt.title('bar plot')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()
```

# Histograms

```python
a= np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a,bins=12)
plt.title('histogram')
plt.show
```

# Box plot using matplotlib

```python
data= [np.random.normal(0,std,100) for std in range(1,4)]

## ractangular box plot
plt.boxplot(data,vert= True, patch_artist= False)
```

# Pie chart

```python
# data to plot
labels= 'python','R','SPSS','Minitab'
sizes=[250,200,360,137]
colors= ['red','blue','orange','gray']
explode=(0.1,0,0,0) # explode 1st slice (i.e here python far from other by 0.1)

#plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',shadow=True)
plt.axis('equal')
plt.show()
```
